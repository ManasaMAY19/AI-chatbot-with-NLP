```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Assuming 'df' is your DataFrame and 'X' and 'Y' are derived from it
df = pd.read_csv('/content/drive/MyDrive/Datasets.ipynb/Salary_Data.csv')
X = df[['Years of Experience']].values
Y = df['Salary'].values

# Drop rows with NaN values in either X or Y
df_cleaned = df.dropna(subset=['Years of Experience', 'Salary'])

# Re-create X and Y from the cleaned DataFrame
X_cleaned = df_cleaned[['Years of Experience']].values
Y_cleaned = df_cleaned['Salary'].values

# Initialize and train the model
model = LinearRegression()
model.fit(X_cleaned, Y_cleaned)
```

```python

import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression

# Assuming 'df' is your DataFrame and 'X' and 'Y' are derived from it
df = pd.read_csv('/content/drive/MyDrive/Datasets.ipynb/Salary_Data.csv')
X = df[['Years of Experience']].values
Y = df['Salary'].values

# Initialize an imputer (e.g., using the mean strategy)
imputer = SimpleImputer(strategy='mean')

# Fit the imputer on your data and transform it
X_imputed = imputer.fit_transform(X)
Y_imputed = imputer.fit_transform(Y.reshape(-1, 1)).flatten()  # Convert Y to a 1D arrayy
# Initialize and train the model
model = LinearRegression()
model.fit(X_imputed, Y_imputed)
```

```python
X_imputed = imputer.fit_transform(X)
model.fit(X_imputed, Y_imputed)
```
